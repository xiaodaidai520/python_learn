#### 1. 哪一次拍卖的中标率首次小于 5%？

```python
#解题思路：
##1.首先求出中标率小于5%的拍卖
##2.再比较中标小于5%的拍卖哪次最早
#考察点：
##创建列，索引
df['percent']= df['Total number of license issued'] / df['Total number of applicants']
df_1=df[df['percent']<0.05].head(20)
df_1
```

#### 3.将第一列时间列拆分成两个列，一列为年份（格式为 20××），另一列为月份（英语缩写），添加到列表作为第一第二列，并将原表第一列删除，其他列依次向后顺延。

```python
#解题思路：
##先把时间拆分成两列，然后再删除原来的列
#考察点：
##拆分列
##第一步，拆分出年和月的一列
df['年份']=df['Date'].apply(lambda x :int(x.split('-')[0]))+2000
df['年份'].head(6)
```

```python
df['月份']=df['Date'].apply(lambda x :x.split('-')[1])
df['月份'].head(6)
```

```python
###### 这个地方再考虑下其他方法
new = ['年份','月份']+list(df.columns[1:-2])
df_newtable = df.reindex(columns=new).copy()
df_newtable.head(6)
```

#### 2. 按年统计拍卖最低价的下列统计量：最大值、均值、0.75 分位数，要求显示在同一张表上。

```python
##### 解题思路：
###按年分组，再统计
#考察点：
##创建列，索引，分组，统计量，分列
groupby_year = df_newtable.groupby('年份')
new = pd.DataFrame()
new['max'] = groupby_year["lowest price "].max()
new['mean'] = groupby_year["lowest price "].mean()
new['075'] = groupby_year["lowest price "].quantile(0.75)
new.head(3)
```

#### 4.现在将表格行索引设为多级索引，外层为年份，内层为原表格第二至第五列的变量名，列索引为月份。

```python
#解题思路：
## 以年份和第二至第五列做行，月份为列
#考察点：
## 变形函数
df_melt =df_newtable.melt(id_vars=['年份', '月份'], var_name='myVariable', value_name='myValue')
df_melt.head(6)
```

```python
result = pd.pivot_table(df_melt,index=['年份','myVariable']
                 ,columns='月份'
                 ,values='myValue')
result.head(10)
```

#### 5.一般而言某个月最低价与上月最低价的差额，会与该月均值与上月均值的差额具有相同的正负号，哪些拍卖时间不具有这个特点？

```python
#解题思路：
##上月最低价与本月最低价差额，再求出上月均值与本月均值的差额，然后两个差额相乘小于0的月份
#考察点：
##统计量，分组，合并，索引排序
df_diff = df[['Date',"lowest price ",'avg price']]

print('最低价、均值与上月差额不同号的月份有：')
for index in range(1, len(df_diff)):
    if ((df_diff.iloc[index]["lowest price "] - df_diff.iloc[index-1]["lowest price "])\
    * (df_diff.iloc[index]['avg price'] - df_diff.iloc[index-1]['avg price'])) < 0:
        print(df_diff.iloc[index]['Date'])
```

#### 6.将某一个月牌照发行量与其前两个月发行量均值的差额定义为发行增益，最初的两个月用 0 填充，求发行增益极值出现的时间。

```python
#解题思路：
##先计算每个月的发行增益，再计算发行增益的最大值和最小值
#考察点：
###考察点：统计量，索引，排序，差分
df_20 = df[['Date','Total number of license issued']]

add = [0,0]
for index in range(2, len(df_20)):
    add.append(df.loc[index]['Total number of license issued'] - np.mean([df.loc[index-1]['Total number of license issued'], df.loc[index-2]['Total number of license issued']]))
df_20['add'] = add
print('极大值：')
print(df_20[df_20['add'] == df_20['add'].max()]['Date'].values)
print('极小值：')
print(df_20[df_20['add'] == df_20['add'].min()]['Date'].values)
```
